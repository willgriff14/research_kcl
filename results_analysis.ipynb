```python


# Data
L = np.array([8, 9, 10, 11, 12])
xi_epoch_6400 = np.array([0.6687, 0.7430, 0.8293, 0.8442, 0.8918])
xi_epoch_3200 = np.array([0.5998, 0.6617, 0.7530, 0.8007, 0.8324])

# Calculate linear fits
fit_6400 = np.polyfit(L, xi_epoch_6400, 1)
fit_3200 = np.polyfit(L, xi_epoch_3200, 1)

# Generate linear fit lines
fit_line_6400 = np.polyval(fit_6400, L)
fit_line_3200 = np.polyval(fit_3200, L)

# Plotting the data with linear fit lines
plt.plot(L, xi_epoch_6400, 'o', label='6400 epoch')
plt.plot(L, fit_line_6400, label=f'6400 epoch fit: y={fit_6400[0]:.4f}x+{fit_6400[1]:.4f}')

plt.plot(L, xi_epoch_3200, 'o', label='3200 epoch')
plt.plot(L, fit_line_3200, label=f'3200 epoch fit: y={fit_3200[0]:.4f}x+{fit_3200[1]:.4f}')

# Adding labels and legend
plt.xlabel('Lattice Size (L)')
plt.ylabel('Correlation Length ()')
plt.title('Correlation Length vs Lattice Size with Linear Fit')
plt.legend()

plt.savefig('corr_length_lattice_size.png')
# Show the plot
plt.show()
```
