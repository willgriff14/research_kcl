```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Will search for correlation length asymptotes on incremental lattice sizes $L=8 \times 8$, $L=9 \times 9$, $L=10 \times 10$, $L=11 \times 11$, $L=12 \times 12$ etc. Then will plot asmptotic correlation lengths against $L$

# $L=10 \times 10:$

```python
L10_epochs = [100, 200, 400, 800, 1600, 3200, 6400 ]
L10_xi = [33.86, 30.55, 22.57, 15.10, 15.33, 15.72,  16.24]
```

```python
L10_epochs = [100,200,400,800,1600,3200,6400]
L10_xi = [7.55, 5.37, ]
```

```python

# Creating a larger figure
plt.figure(figsize=(6, 4))

# Creating a scatter plot with larger markers
plt.scatter(L10_epochs, L10_xi, color='blue', s=50, alpha=0.7, marker='x')  # 's' is the size of markers

# Adding grid
plt.grid(True, which='both', linestyle='--', linewidth=0.5)


# Adding title and labels with increased font size
plt.title(r'$WW^T$ Correlation Length vs. Training Epochs for $L=10x10$', fontsize=10)
plt.xlabel('Training Epochs', fontsize=8)
plt.ylabel(r'Correlation Length $\xi$', fontsize=8)



plt.xlim([0, max(L10_epochs) + 50])  # Adjust as needed
plt.ylim([0, max(L10_xi) + 1])       # Adjust as needed


# Display the plot
plt.show()
```
