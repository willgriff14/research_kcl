# Generating spin configurations via Monte Carlo - $L=9$

```python
import pandas as pd

import sys
# sys.path.append('../modules')
from mc_ising2d import IsingMC
```

```python
L = 9
```

```python
T_min = 0.0
T_max = 6.0
num_T = 25
MC_states_per_T = 2000

MC_sweeps_per_state= 100
MC_eq_sweeps= 50
```

```python
ising2D = IsingMC(L= L, 
                  T_min= T_min, 
                  T_max= T_max, 
                  num_T= num_T, 
                  MC_states_per_T= MC_states_per_T,
                  MC_sweeps_per_state= MC_sweeps_per_state,
                  MC_eq_sweeps= MC_eq_sweeps,
                  include_Tc= False)
```

```python
ising2D.fit(thermodynamics = True)
```

```python
ising2D.plot_thermodynamics()
# this shows specific heat (bottom left) and suseptibility (bottom right) per spin
```

```python
spin_MC = ising2D.thermodymanics_MC_df()
```

```python
spin_MC.to_pickle(f'L{L}_Ising2D_MC.pkl')
```
