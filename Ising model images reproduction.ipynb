```python
import numpy as np
import matplotlib.pyplot as plt
```

# 2D Ising model on 100x100 square lattice at criticality 

```python
# Parameters
L = 100  # size of lattice
T = 2 / np.log(1 + np.sqrt(2))  # temperature
J = 1.0  # interaction strength
num_steps = 1000  # number of Monte Carlo steps

# Initialize lattice
lattice = np.ones((L, L))

# Function to calculate the energy change for a single spin flip
def deltaE(i, j):
    return 2 * J * lattice[i, j] * (lattice[(i-1)%L, j] + lattice[(i+1)%L, j] +
                                    lattice[i, (j-1)%L] + lattice[i, (j+1)%L])

# Metropolis update rule
for _ in range(num_steps):
    for _ in range(L * L):
        i, j = np.random.randint(0, L, 2)
        dE = deltaE(i, j)
        if dE < 0 or np.random.rand() < np.exp(-dE / T):
            lattice[i, j] *= -1

# Display the final state
plt.imshow(lattice, cmap='binary')
plt.show()
```

# Comparing $T=T_c$ , $T>T_c$ and $T< T_c$ on 500x500 lattices

```python
# Parameters
L = 500  # size of lattice
J = 1.0  # interaction strength
num_steps = 1000  # number of Monte Carlo steps

# Temperatures
T_below = 2.0
T_critical = 2 / np.log(1 + np.sqrt(2))  # critical temperature
T_above = 2.5

temperatures = [T_below, T_critical, T_above]

# Function to calculate the energy change for a single spin flip
def deltaE(i, j):
    return 2 * J * lattice[i, j] * (lattice[(i-1)%L, j] + lattice[(i+1)%L, j] +
                                    lattice[i, (j-1)%L] + lattice[i, (j+1)%L])

# Perform simulation for each temperature
for T in temperatures:
    # Initialize lattice
    lattice = np.ones((L, L))

    # Metropolis update rule
    for _ in range(num_steps):
        for _ in range(L * L):
            i, j = np.random.randint(0, L, 2)
            dE = deltaE(i, j)
            if dE < 0 or np.random.rand() < np.exp(-dE / T):
                lattice[i, j] *= -1

    # Display the final state
    plt.figure(figsize=(10, 10))
    plt.imshow(lattice, cmap='binary')
    plt.title(f'T = {T:.3f}')
    plt.show()
```
